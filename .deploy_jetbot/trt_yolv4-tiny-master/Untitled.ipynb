```python
import cv2

# 1. 載入模型 (OpenCV 直接讀取)
net = cv2.dnn.readNetFromONNX("best.onnx")

# 2. 如果你的 Jetbot 有 GPU，強制 OpenCV 使用 CUDA
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)

# 3. 推論測試 (不需要安裝 ultralytics)
def detect(frame):
    blob = cv2.dnn.blobFromImage(frame, 1/255.0, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)
    outputs = net.forward()
    return outputs
```

```python
import onnxruntime as ort
import numpy as np

# 1. 載入模型 (這個套件的相容性極高，不會出現 OpenCV 那種解析錯誤)
session = ort.InferenceSession("best.onnx")

# 2. 查看輸入層名稱 (這是為了寫後續的預處理)
input_name = session.get_inputs()[0].name
print(f"✅ 模型載入成功！輸入層名稱: {input_name}")

# 3. 測試推論 (送入一張全黑的假圖片來確認模型能運作)
dummy_input = np.zeros((1, 3, 416, 416), dtype=np.float32)
outputs = session.run(None, {input_name: dummy_input})
print("✅ 模型推論測試成功！")
```
